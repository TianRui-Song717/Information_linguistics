### 读取name列并放入将相关数据放入final_df中

```python
sorted_cleaned_data.drop(columns=["new_index","old_index"], inplace=True)
```

```python
sorted_cleaned_data = sorted_cleaned_data.reset_index()
```

```python
sorted_cleaned_data.head()
```

```python
def col_name_to_zishun_df(x):
    dict_1 = {}
    dict_1["name"] = x["name"]
    dict_1["D"] = x["baike_nickname"]
    dict_1["Y"] = ""
    dict_1["S"] = x["class"]
    dict_1["F"] = ""
    return dict_1
```

```python
all_col_name = []
sorted_cleaned_data.apply(lambda x: all_col_name.append(col_name_to_zishun_df(x)), axis=1)
```

### 放入最终导出的df中

# 将final_df清洗+分行

```python
final_df["name"].apply(print)
```

```python
final_cleaned = final_df[final_df["name"] != "None"]
cleaned_final = final_cleaned[final_cleaned["name"] != 0]
cleaned_final
```

```python
final = cleaned_final.reset_index().drop(columns="index")
```

清洗完成

```python
linshi_df = final.copy()
def informal_to_single_line(x):
    informal_name_list = x["name"].split("、")
    this_formal_list = []
    for each_informal_name in informal_name_list:
        this_dict = {"name":each_informal_name}
        this_dict["Y"] = x["Y"]
        this_dict["D"] = ""
        this_dict["S"] = ""
        this_dict["F"] = ""
        this_formal_list.append(this_dict)
    return this_formal_list
# linshi_df[final["name"].apply(lambda x: True if '、' in x else False)]
temp_df = final[final["name"].apply(lambda x: True if '、' in x else False)]
```

```python
all_informal_name_list = []
temp_df.apply(lambda x: all_informal_name_list.append(informal_to_single_line(x)),axis=1)
```

```python
# linshi_df = linshi_df[final["name"].apply(lambda x: True if '、' not in x else False)]
linshi_df = pd.DataFrame(columns=["name","Y","D","S","F"])
```

```python
final_informal_list = []
for each_list in all_informal_name_list:
    for each_informal_name in each_list:
        final_informal_list.append(each_informal_name)
```

```python
linshi_df = linshi_df.append(final_informal_list, ignore_index=True)
linshi_df
```

```python
linshi_df.replace(to_replace="None", value="", inplace=True)
```

分行完成

```python
formal_name_df = final[final["name"].apply(lambda x: True if '、' not in x else False)]
formal_name_df
```

```python
informal_name_df = linshi_df
```

```python
txt_data_df = pd.concat([formal_name_df,informal_name_df], axis=0)
```

```python
sorted_txt_data = pd.read_excel("字顺表.xlsx")
sorted_txt_data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
sorted_txt_data.fillna(value="", inplace=True)
```

```python
#写入txt文件
def output_to_list(x,file):
    file.writelines(x["name"])
    if (x["Y"] != ""):
        file.writelines(f'\n\tY\t{x["Y"]}')
    if (x["D"] != ""):
        if('、' in x["D"]):
            informal_name_list = x["D"].split("、")
            first_D = 1
            for each_informal_name in informal_name_list:
                if first_D == 1:
                    file.writelines(f'\n\tD\t{each_informal_name}')
                else:
                    file.writelines(f'\n\t \t{each_informal_name}')
                first_D += 1
        else:
            file.writelines(f'\n\tD\t{x["D"]}')
    if (x["S"] != ""):
        file.writelines(f'\n\tS\t{x["S"]}')
    if (x["F"] != ""):
        file.writelines(f'\n\tF\t{x["F"]}')
    file.writelines("\n")

        
with open("字顺表.txt", 'wt',encoding='utf-8') as file:
    sorted_txt_data.apply(lambda x:output_to_list(x, file), axis=1)
```
