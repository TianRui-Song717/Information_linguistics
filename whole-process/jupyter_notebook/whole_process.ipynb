### *爬取页面的函数*

```python
def Get_html(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return "Error"
```

```python
html = Get_html("http://www.boohee.com/food/search?keyword=猪血")
```

- real

```python
# 开始爬取时记得print提示信息
# 爬取页面的函数
def get_food_info(search_url,name):
    base_url = "http://www.boohee.com"
    food_informal_name = "None"
    food_class = "None"
    food_nutrient = "None"
    food_energy = "None"
    food_carbohydrates = "None"
    food_fat = "None"
    food_protein = "None"
    food_cellulose = "None"
    try:
        html = Get_html(search_url)
        # 获取该食物的信息页面链接
        soup = BeautifulSoup(html, "html.parser")
        food_list = soup.find("ul",{"class":"food-list"})
        food_item = food_list.find_all("li")[0]
        food_item_link_add = food_item.find("div",{"class":"text-box"}).h4.a["href"]
        food_item_link = base_url + food_item_link_add
        # 获取该食物的详细信息
        food_detail_html = Get_html(food_item_link)
        detail_soup = BeautifulSoup(food_detail_html,"html.parser")
        # 别名（同义词）
        if(detail_soup.find("ul",{"class":"basic-infor"}).find_all("li")[0].b.text == "别名："):
            food_informal_name = detail_soup.find("ul",{"class":"basic-infor"}).find_all("li")[0].text[3:]
        # 所属类别
        if(detail_soup.find("ul",{"class":"basic-infor"}).find_all("li")[2].b.text == "分类："):
            food_class =  detail_soup.find("ul",{"class":"basic-infor"}).find_all("li")[2].text[3:]
        else:
            food_class =  detail_soup.find("ul",{"class":"basic-infor"}).find_all("li")[1].text[3:]
        # 营养素
        food_nutrient = detail_soup.find("div",{"class":"nutr-tag"}).div.find_all("dl")
        food_energy = food_nutrient[1].find_all("dd")[0].find("span",{"class":"dd"}).span.text # 能量
        food_carbohydrates = food_nutrient[1].find_all("dd")[1].find("span",{"class":"dd"}).text # 碳水
        food_fat = food_nutrient[2].find_all("dd")[0].find("span",{"class":"dd"}).text # 脂肪
        food_protein = food_nutrient[2].find_all("dd")[1].find("span",{"class":"dd"}).text # 蛋白质
        food_cellulose = food_nutrient[3].find_all("dd")[0].find("span",{"class":"dd"}).text # 纤维素
    except:
        print(f"*****Something wrong with {name} food!*****")
    #返回一个字典
    return {"name":name, "informal_name":food_informal_name, "class":food_class, "energy":food_energy, "carbohydrates":food_carbohydrates, "fat":food_fat, "protein":food_protein, "cellulose":food_cellulose}
```

```python
get_food_info("http://www.boohee.com/food/search?keyword=火腿", "火腿")
```

### 将所有页面全部爬取（从259开始）

```python
# https://fq.chinafcd.org/foodinfo/259.html
no_time = 0
df = pd.DataFrame(columns=["name","class","subclass"])
for i in range(259,1900):
    print(f"-- get NO.{i} food!")
    html = Get_html(f"https://fq.chinafcd.org/foodinfo/{i}.html")
    if(html == "Error"):
        no_time += 1
        if (no_time == 3):
            break;
    else:
        no_time = 0
        soup = BeautifulSoup(html,"html.parser")
        food_name = soup.find("div",{"class":"food_introduce_top"}).h1.text
        food_class = soup.find("div",{"class":"food_introduce_top"}).find_all("span")[0].text[4:]
        food_subclass = soup.find("div",{"class":"food_introduce_top"}).find_all("span")[1].text[6:]
        df = df.append({"name":food_name, "class": food_class, "subclass": food_subclass}, ignore_index=True)
    print(f"- get NO.{i} food finished!")
df.to_csv("all_national_food_info.csv", encoding="utf_8_sig")
```
